```python
# Importing Libraries
import pandas as pd
from sklearn import preprocessing 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
# Import Train data
train_data = pd.read_csv('SalaryData_train.csv')
train_data
```

```python
train_data.info()
```

```python
train_data.Salary.unique()
```

```python
cols = train_data.columns.drop(['age','educationno','capitalgain', 'capitalloss', 'hoursperweek'])
cols
```

```python
# Encoding the categorical data
encoded_train_data = train_data[['age','educationno','capitalgain', 'capitalloss', 'hoursperweek']]
for column in cols:
    le = preprocessing.LabelEncoder()
    encoded_train_data[f'encoded_{column}'] = le.fit_transform(train_data[column])
```

```python
encoded_train_data
```

```python
X_train = encoded_train_data.iloc[:,:13]
Y_train = encoded_train_data.iloc[:,13]
```

```python
X_train
```

```python
Y_train
```

```python
# Creating Model
model = GaussianNB()
model.fit(X_train,Y_train)
```

```python
test_data = pd.read_csv('SalaryData_Test.csv')
```

```python
# Encoding Test Categorical data
encoded_test_data = test_data[['age','educationno','capitalgain', 'capitalloss', 'hoursperweek']]
cols_test = test_data.columns.drop(['age','educationno','capitalgain', 'capitalloss', 'hoursperweek'])
for column in cols_test:
    le = preprocessing.LabelEncoder()
    encoded_test_data[f'encoded_{column}'] = le.fit_transform(test_data[column])
```

```python
encoded_test_data
```

```python
X_test = encoded_test_data.iloc[:,:13]
Y_test = encoded_test_data.iloc[:,13]
```

```python
X_test
```

```python
Y_test
```

```python
Y_predicted = model.predict(X_test)
Y_predicted
```

```python
# Confusion Metrics
confusion_matrix(Y_test,Y_predicted)
```
